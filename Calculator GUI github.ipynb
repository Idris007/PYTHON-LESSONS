# Simple calculator GUI

```python
from tkinter import *
import os


root = Tk()
# set title for your window widget
root.title('Demo Simple Calculator')
# adjust the length and bredth of window widget
root.geometry("320x340")
# change default tkinter icon into your own
root.iconbitmap(os.path.join(os.getcwd(), 'Calculator.ico'))
# create entry widget
equation = StringVar()
e = Entry(root, width= 50, borderwidth= 10, fg='black', bg='blue', textvariable=equation)
# insert default instruction on entry widget
equation.set('Enter any number')
e.grid(row= 1, column=0 ,columnspan=3)


# function to check for decimal point in entry numbers
def check_num(num):
    global c_num
    if '.' in c_num:
        c_num = float(num)
    else:
        c_num = int(num)



# function to have multiple numbers displayed on the entry box
def click(new_number):
    current_number = e.get()
    e.delete(0, END)
    e.insert(0, str(current_number) + str(new_number))


# function for mathematical operation of multiplication
def multiply():
    current_number = e.get()
    e.delete(0, END)
    global c_num
    c_num = current_number
    global math
    math = 'multiplication'
    # smart function checks if decimal in entry and assigns it to a class int() or float()
    check_num(c_num)




# function for mathematical operation of addition
def add():
    current_number = e.get()
    global c_num
    global math
    math = 'addition'
    c_num = current_number
    e.delete(0, END)
    # smart function checks if decimal in entry and assigns it to a class int() or float()
    check_num(c_num)



# function for mathematical operation of subtraction
def subtract():
    current_number = e.get()
    global c_num
    c_num = current_number
    global math
    math = 'subtraction'
    e.delete(0, END)
    # smart function checks if decimal in entry and assigns it to a class int() or float()
    check_num(c_num)



# function for mathematical operation of division
def divide():
    current_number = e.get()
    e.delete(0, END)
    global c_num
    c_num = current_number
    global math
    math = 'division'
    # smart function checks if decimal in entry and assigns it to a class int() or float()
    check_num(c_num)



# function to collect second number and process the mathematical operation on both numbers
def equal():
    second_number = e.get()
    e.delete(0, END)
    # condition of math equal to multiplication is met when the function multiply() is used
    if math == 'multiplication':
        if '.' in second_number:
            result = c_num * float(second_number)
        else:
            result = c_num * int(second_number)
        e.insert(0, result)
    # condition of math equal to division is met when the function divide() is used
    if math == 'division':
        if '.' in second_number:
            result = c_num / float(second_number)
        else:
            result = c_num / int(second_number)
        e.insert(0, result)
    # condition of math equal to addition is met when the function add() is used
    if math == 'addition':
        if '.' in second_number:
            result = c_num + float(second_number)
        else:
            result = c_num + int(second_number)
        e.insert(0, result)
    # condition of math equal to subtraction is met when the function subtract() is used
    if math == 'subtraction':
        if '.' in second_number:
            result = c_num - float(second_number)
        else:
            result = c_num - int(second_number)
        e.insert(0, result)



# function to clear entry box
def clear():
    e.delete(0, END)



# buttons styling and positioning on widget/window
button_1 = Button(root, text='1', command=lambda: click(1), bg='red', padx=40, pady=10, borderwidth=5)
button_2 = Button(root, text='2', command=lambda: click(2), bg='red', padx=40, pady=10, borderwidth=5)
button_3 = Button(root, text='3', command=lambda: click(3), bg='red', padx=40, pady=10, borderwidth=5)

button_4 = Button(root, text='4', command=lambda: click(4), bg='red', padx=40, pady=10, borderwidth=5)
button_5 = Button(root, text='5', command=lambda: click(5), bg='red', padx=40, pady=10, borderwidth=5)
button_6 = Button(root, text='6', command=lambda: click(6), bg='red', padx=40, pady=10, borderwidth=5)

button_7 = Button(root, text='7', command=lambda: click(7), bg='red', padx=40, pady=10, borderwidth=5)
button_8 = Button(root, text='8', command=lambda: click(8), bg='red', padx=40, pady=10, borderwidth=5)
button_9 = Button(root, text='9', command=lambda: click(9), bg='red', padx=40, pady=10, borderwidth=5)

button_0 = Button(root, text='0', command=lambda: click(0), bg='red', padx=40, pady=10, borderwidth=5)
button_add = Button(root, text='+', command=add, bg='yellow', padx=15, pady=10, borderwidth=5)
button_multiply = Button(root, text='*', command=multiply, bg='yellow', padx=40, pady=10, borderwidth=5)
button_divide = Button(root, text='/', command=divide, bg='yellow', padx=15, pady=10, borderwidth=5)
button_equal = Button(root, text='=', command=equal, bg='grey', padx=40, pady=10, borderwidth=5)

button_subtract = Button(root, text='-', command=subtract, bg='yellow', padx=40, pady=10, borderwidth=5)
button_clear = Button(root, text='clear', command=clear, bg='yellow', padx=30, pady=10, borderwidth=5)
button_dot = Button(root, text='.', command=lambda: click('.'), bg='grey', padx=40, pady=10, borderwidth=5)



# grid() method for arrangement of buttons, other method pack() can be used
button_7.grid(row=2, column=0)
button_8.grid(row=2, column=1)
button_9.grid(row=2, column=2)

button_4.grid(row=3, column=0)
button_5.grid(row=3, column=1)
button_6.grid(row=3, column=2)

button_1.grid(row=4, column=0)
button_2.grid(row=4, column=1)
button_3.grid(row=4, column=2)

button_add.grid(row=5, column=1, sticky= E)
button_divide.grid(row=5, column=1, sticky=W)
button_multiply.grid(row=5, column=2)
button_0.grid(row=5, column=0)
button_equal.grid(row=6, column=2)

button_subtract.grid(row=6, column=0)
button_clear.grid(row=6, column=1)
button_dot.grid(row=7, column=0)



# The graphical user interface is in a constant loop
root.mainloop()
```
