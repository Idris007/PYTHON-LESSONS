```python
def two_digit():
    x = int(input('Enter 1st digit...:'))
    y = int(input('Enter 2nd digit...:'))
    operation= input('Enter operation e.g *, +, -, /...:')
    if operation == '*':
        multiply_2(x, y)
    elif operation == '+':
        add_2(x, y)
    elif operation == '/':
        divide_2(x, y)
    elif operation == '-':
        minus_2(x, y)


def three_digit():
    x = int(input('Enter 1st digit...:'))
    y = int(input('Enter 2nd digit...:'))
    z = int(input('Enter 3rd digit...:'))
    operation= input('Enter operation e.g *, +, -, /...:')
    if operation == '*':
        multiply_3(x, y, z)
    elif operation == '+':
        add_3(x, y, z)
    elif operation == '/':
        divide_3(x, y, z)
    elif operation == '-':
        minus_3(x, y, z)


def multiply_2(x, y):
    z = x * y
    print(z)

def multiply_3(x, y, z):
    a = x * y * z
    print(a)

def add_2(x, y):
    z = x + y
    print(z)

def add_3(x, y, z):
    a = x + y + z
    print(a)

def divide_2(x, y):
    z = x / y
    print(z)

def divide_3(x, y, z):
    a = x / y / z
    print(a)

def minus_2(x, y):
    z = x - y
    print(z)

def minus_3(x, y, z):
    a = x - y - z
    print(a)

while True:
    user = input('To calculate 2 or 3 numbers enter \n 1. 2-numbers \n 2. 3-numbers \n:')
    if user == '2':
        three_digit()

    elif user == '1':
        two_digit()
        
    elif user != '1' or user != '2':
        print('Invalid input...')












```
