# Youtube downloader

```python
from tkinter import *
from tkinter import ttk
from pytube import YouTube
from tkinter import messagebox
import time
import os

root = Tk()
# calculates the download percentage of downloading youtube video
def percentageCalculator(x, y):
    #Calculate percentages
    r = x / y * 100
    return r

# function to run download action
def click(progress_bar, status, status1):
    # entry widget 1 to collect url path
    e1 = entry1.get()
    # entry widget 2 to collect download path
    e2 = entry2.get()

    # class YouTube collects url path
    yt = YouTube(e1)
    # selcting the first stream video
    video = yt.streams.filter(progressive=True, file_extension='mp4').first()
    # get the size of stream in byte and convert it into megabytes
    size_of_file = int(video.filesize/(1024*1024))
    # convert file size into a string format
    size_of_file_string = str(str(round(video.filesize / (1024 * 1024))) + 'MB')
    # using the range function to iterate over the integer of size in megabyte
    alist = range(size_of_file)
    # save directory of the file collected from entry widget 2
    save_path = (e2)
    # download video file and save to path
    video.download(save_path)

    # open a message box
    log = open("log.txt", "a")

    # try block with exception code
    try:
        # create count p to number the iteration of for loop
        p = 0
        # loop to iterate over the range of filesize
        for i in alist:
            # count starts from 1
            p += 1
            # pass argument p and length of alist to function percentageCalculator()
            unit = percentageCalculator(p, len(alist))

            # time sleep as function is iterating
            time.sleep(1)
            # log message on completion
            log.write(str('\n[OK]'))
            # the value of the progress bar will calculated by function percentageCalculator()
            progress_bar['value'] = unit
            # this will output the progress of the size of download file on a label
            status['text'] = f"Working on {i + 1}"
            # this will output the size of the file been downloaded on a label
            status1['text'] = size_of_file_string
            # All this processes will be updated to the root gui
            root.update()
        # information displayed on message box
        messagebox.showinfo('Info', "Process completed!")
    # exception when there is any error
    except Exception as e:
        # in the event of an error display e and close program using sys.exit() function
        messagebox.showinfo('Info', f"ERROR: {e}")
        sys.exit()
    # close the message log
    log.close("log.txt")


# set the background colour of GUI window
root.configure(background="light green")
# title of window widget
root.title('Youtube Downloader')
# window dimension size
root.geometry('280x200')
# icon file path
root.iconbitmap(os.path.join(os.getcwd(), 'youtube.ico'))

# button frame for the progress bar
buttonFrame = LabelFrame(text="Progress Bars", bg='light blue', width=20, borderwidth=5)
buttonFrame.grid(column=0, row=10)

# entry frame for the progress bar
entry = LabelFrame(text="Entries", bg='light green')
entry.grid(column=0, row=5)

# prgress bar inside the button frame
progress_bar = ttk.Progressbar(buttonFrame, length=200, orient='horizontal', mode='determinate')
progress_bar.grid(column=0, row=6, pady=2)

# button to download file
button1 = Button(text="Download", bg='black', fg='white', command=(lambda: click(progress_bar, status, size)))
button1.grid(column=0, row=8)

# entry widget 1
entry1 = Entry(entry, width=35, bg='yellow')
entry1.grid(column=0, row=3, ipadx=30)
entry1.insert(0, 'Enter url path')

# entry widget 2
entry2 = Entry(entry, width=35, bg='yellow')
entry2.grid(column=0, row=4, ipadx=30)
entry2.insert(0, 'Enter download path...')

# status label
status = Label(text="Click button to start process..",bg='light green', relief=SUNKEN, anchor=W, borderwidth=2, width=30)
status.grid(column=0, row=11, columnspan=35, sticky='S')

# size label
size = Label(text="Display download file size..", bg='light green', relief=SUNKEN, anchor=W, borderwidth=2, width=30)
size.grid(column=0, row=12, columnspan=35, sticky='S')


root.mainloop()
```
